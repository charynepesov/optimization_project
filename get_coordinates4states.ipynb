```python
import requests
import csv
```

```python
apikey=''
url = 'https://maps.googleapis.com/maps/api/geocode/json?address='
city = 'Ankara'
```

```python
iller = []
with open('states_of_turkey.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in reader:
        iller.append(row[0])
```

```python
with open('tr_state_coordinates.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile, delimiter=',', quotechar='|', quoting=csv.QUOTE_MINIMAL)
    for il in iller:
        response = requests.get(url+il+'&key='+apikey)
        resp_json_payload = response.json()
        if resp_json_payload['status'] == 'OK':
            lat = resp_json_payload['results'][0]['geometry']['location']['lat']
            lng = resp_json_payload['results'][0]['geometry']['location']['lng']
            print(il+': lat:'+str(lat)+', lng:'+str(lng))
            writer.writerow([il, lat, lng])
```
